```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y)
plt.show()
```

```
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='red',linestyle='-',marker='^',linewidth=2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Line chart")
plt.legend(['my data series'])
plt.grid(False)
plt.show()
```

```
categories=['A','B','C','D']
values=[1,4,9,16]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='orange')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```
# simple data
data=[12,15,13,17,16,16,19,12,14,16,14,17,18,18,15,14]
#create histogram
plt.hist(data,bins=16)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```
