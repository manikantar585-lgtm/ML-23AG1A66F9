**importing dataset**

```python
import pandas as pd
data=pd.read_csv('Name_Age_Gender_Marks_City - Sheet1 (1).csv')
data.head()
```

```python
data.fillna({
    'Age':data['Age'].mean(),
    'Marks':data['Marks'].mean(),
    'City':'Unknown'
},inplace=True)
print(data)
```

```python
import pandas as pd
# create example data including duplicates
data ={
    'Name': ['Mani', 'ajay', 'Mani'],
    'Age': [18, 18, 18],
    'Gender': ['M','M','M'],
    'Marks': [85,92,85]
}
df=pd.DataFrame(data)
print("original data:")
print(df)
#Remove duplicates across all colums
df_no_duplicates=df.drop_duplicates()
print("\nAfter removing duplicates:")
print(df_no_duplicates)
```
